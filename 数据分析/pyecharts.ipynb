### 第一个pyecharts图表
- 熟悉bar方法
- 准备一组学生三次模拟考试成绩
- 对比学生三次模拟考试成绩

```python
subjects = ['数学', '语文', '英语', '文综', '理综']
scores = {'S1':[100, 90, 33, 230, 263], 
          'S2':[88, 120, 100, 260, 242], 
          'S3':[140, 103, 130, 283, 287]}
```

```python
from pyecharts import Bar
```

```python
help(Bar)
```

```python
# 熟悉bar方法
# 创建bar对象，表title
bar = Bar('三次模拟考试', subtitle='XX学校', width='300')
# 使用add方法添加数据，数据描述，x轴数据，y轴数据
for key in scores:
    bar.add(key, subjects, scores[key])
# bar.add('S1', subjects, scores['S1'], is_convert=True)
# bar.add('S2', subjects, scores['S2'], is_convert=True)
# bar.add('S3', subjects, scores['S3'], is_convert=True)
# 图表输出
bar
# 保存图表
bar.render
```
