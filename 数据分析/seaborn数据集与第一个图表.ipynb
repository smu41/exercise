###  使用pointplot绘制图表
1）选择setosa种类，绘制花萼长度分布

2）setosa种类花萼长度与宽度对应点分布（x轴为花萼长度，y轴为宽度）

3）所有种类鸢尾花长度与宽度对应点分布（x轴为花萼长度，y轴为宽度）

```python
# setosa种类，x轴为0~N，y轴为花萼长度
sdata = pddata[pddata['species'] == 'setosa']
sdata
```

```python
sepalData = sdata['sepal_length']
sepalData_width = sdata['sepal_width']
```

```python
plt.figure(figsize=(15,4))
sns.set(style='darkgrid', palette='deep')
sns.pointplot(x=sepalData.index.values, y=sepalData.values, color='r')
sns.pointplot(x=sepalData_width.index.values, y=sepalData_width.values, color='g')
```

```python
# setosa种类花萼长度与宽度对应点分布
sns.pointplot(x='sepal_width', y='sepal_length', data = sdata, color='g')
```

```python
# 所有种类鸢尾花花萼长度与宽度对应点分布
plt.figure(figsize=(15,5))
sns.pointplot(x='sepal_length', y='sepal_width', hue='species', data=pddata)
```

```python
pddata
```

```python
plt.figure(figsize=(15,5))
sns.pointplot(x='petal_length', y='petal_width', hue='species', data=pddata)
```
