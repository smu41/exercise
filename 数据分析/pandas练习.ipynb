### aggregate方法：

```python
# 计算最大值、最小值、平均值、个数（mean, min, max, size）
# obj.aggregate(['min', 'max', 'mean'])
obj['rating'].aggregate(['min', 'max'])
```

```python
obj.aggregate({'rating': [max, min], 'title': 'size'})
```

### filter方法：

```python
# 过滤评分均值大于4的所有电影
count = 0
def func(data):
    global count
    count += 1
    return data['rating'].mean() >= 4
r = obj.filter(func)
```

```python
r.groupby('title').aggregate({'rating': 'mean'})
```

### transform方法：

```python
# 计算每一个评分与评分平均值差：
def func(data):
    return data - data.mean()
t.groupby('movieId')['rating'].transform(func)
```

### apply方法：

```python
def func(data):
    return data['rating'] - data['rating'].max()
obj.apply(func)
```

### 数据处理，评分求均值
#### 统计评级人数超过20的所有电影，计算平均分，根据平均分进行统计（0-1，1-2..4-5）

```python
# 根据ID分组
# 过滤评价人数小于20
obj = df.groupby('movieId')
tmp = obj.filter(lambda x: len(x) > 20)
```

```python
# 求均值
obj20 = tmp.groupby('movieId')
t_means = obj20.apply(lambda x: x['rating'].mean())
t_means
```

 ### 分组处理，pd.cut

```python
help(pd.cut)
```

```python
tmp = pd.cut(t_means, range(0, 6), right=False)
```

```python
tmpdf = pd.DataFrame(tmp, columns=['levels'])
result = tmpdf.groupby('levels').size()
```
