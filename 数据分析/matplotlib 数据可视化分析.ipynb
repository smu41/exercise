```python
x = np.linspace(0, 20, num=100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1)
plt.plot(x, y2)
plt.title('sin(x) and cos(x)')
plt.grid(True)
```

###  添加图例

```python
plt.plot(x, y1, '-y^', x, y2, ':r<')
plt.legend(['sin(x)', 'cos(x)'])
plt.show()
```

```python
lines = plt.plot(x, y1, '-y^', x, y2, ':r<')
plt.legend(lines, ['sin(x)-', 'cos(x).'])
plt.show()
```

```python
help(plt.legend)
```

```python
lines = plt.plot(x, y1, '-y^', x, y2, ':r<')
plt.legend(lines, ['sin(x)-', 'cos(x).'], loc = 2)
plt.show()
```

```python
lines = plt.plot(x, y1, '-y^', x, y2, ':r<')
plt.legend(lines, ['sin(x)-', 'cos(x).'], loc = 2, shadow=True)
plt.show()
```

```python
lines = plt.plot(x, y1, '-y^', x, y2, ':r<')
leg = plt.legend(lines, ['sin(x)-', 'cos(x).'], loc = 2)
leg.get_frame().set_color('g')
leg.get_frame().set_alpha(0.5)
leg.set_title('x-y')
plt.show()
```

###  设置xy轴的值

```python
x = np.linspace(0, 20, 100)
y = np.sin(x)
plt.plot(x, y)
plt.xlim(0, 5)
plt.ylim(-1.5, 1.5)
plt.xticks([0,1,2,3,4,5], ['a','b','c','d','e'], rotation=180)
```

###  子图

```python
plt.subplot(1,2,1)
y = np.sin(x)
plt.plot(x, y)
plt.title('Sin(x)')
plt.subplot(1,2,2)
y = np.cos(x)
plt.plot(x, y)
plt.title('Cos(x)')
```

```python
help(plt.subplot)
```

```python
plt.subplot(211)
y = np.sin(x)
plt.plot(x, y)
plt.title('Sin(x)')
plt.subplot(212, facecolor='y')
y = np.cos(x)
plt.plot(x, y)
plt.title('Cos(x)')
plt.subplots_adjust(hspace = 0.5)
```
