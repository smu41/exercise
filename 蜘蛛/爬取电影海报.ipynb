```python
url = 'https://movie.douban.com/top250?start=0&filter='
req = request.urlopen(url)
content = req.read().decode('utf-8')
```

```python
span = re.findall(r'<span class="next".*?</span>', content, re.S)
span = span[0]
span
```

```python
href = re.search(r'href="(.*)"', span)
href.groups()
```

```python
urlhead = 'https://movie.douban.com/top250'
url = urlhead + href.groups()[0]
url
```

```python
url = 'https://movie.douban.com/top250?start=0&filter='
while True:
    req = request.urlopen(url)
    content = req.read().decode('utf-8')
    span = re.findall(r'<span class="next".*?</span>', content, re.S)
    span = span[0]
    href = re.search(r'href="(.*)"', span)
    if not href:
        break
    url = urlhead + href.groups()[0]
    print(url)
```

```python
img = tag.find('img')
```

```python
listdiv = obj.find_all('div', class_ = 'item')
```

```python
for div in listdiv:
    img = div.find('img')
    print(img.get('alt'), img.get('src'))
```
