### 3. 案例：定制天气预报

```python
import requests  # pip install requests
url = 'https://www.sojson.com/open/api/weather/json.shtml?city='
city = '北京'
ret = requests.get(url + city)
print(ret.json())
```

```python
# 把返回的json写到json文件，weather.json
# 用的时候从文件读取
# 像字典一样取值
d = ret.json()
# print(d['status'])
# print(d['city'])
# print(d['data'])
# print(d['data']['yesterday'])

def cool_weather(data):
    """定制化天气预报"""
    weather_list = data['data']['forecast']
#     print(weather_list)
    for day in weather_list:
        print(day['date'], day['high'], day['fx'], day['type'])
    
cool_weather(d)
```

### 4. 案例：博客词频统计

目的： 自定义一个API，能让别人用我们的函数功能

```python
from bs4 import BeautifulSoup  # pip install bs4
```

```python
blog_url = 'http://blog.51cto.com/de8ug/2110764'
```

```python
data = requests.get(blog_url)
print(data)
# print(data.text)
```

```python
contents = BeautifulSoup(data.text, 'html.parser')
# print(contents)
```

```python
all_p = contents.find_all('p')
# print(all_p)
all_text = ''
for p in all_p:
    print(p.text)
    all_text += str(p.text)  # 确认都是字符串
```

```python
import jieba  # pip install jieba
```

```python
text = jieba.cut(all_text)
text_list = []
for t in text:
#     print(t)
    text_list.append(t)
```

```python
import collections

# new_list = list(text_list)  # 如果不是列表需要先转成列表
count = collections.Counter(text_list)
for key, val in count.most_common(30):  # 有序（返回前n个）
    print(key, val)
```

```python
import collections

def get_most_common(text_list, max_num=30):
    """根据max_num取排名靠前的词和出现次数"""
    ret = {'status': 0, 'statusText': 'ok', 'data': {}}  # 返回提示格式, status（状态码），statustext（状态值）
    try:
        new_list = list(text_list)
        count = collections.Counter(new_list)
        ret['data'] = count.most_common(30)  # 有序，返回前n个
    except Exception as e:
            ret['status'] = 1
            ret['statusText'] = e  # TODO
    return ret

get_most_common(text_list)
```
