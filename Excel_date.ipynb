```python
class Excel_data:
    def __init__(self, From, to, sht_name):
        self.From = From
        self.to = to
        self.sht_name = sht_name
        
    def run_excel(self):
        self.data_path()
        self.read_data('name')
        self.add_sheet()
        self.add_header()
        self.data_screen()
        self.save()
        
    def data_path(self):
        print('请输入要读去的数据路径：')
        From = input()
        os.chdir(From)
    
    def read_data(self, name):
        print('请输入文件名：')
        name = input()
        wb = openpyxl.load_workbook(name)
        
    def add_sheet(self):
        print('请输入工作表名：')
        self.sht_name = input()
        wb.create_sheet(self.sht_name)
        
    def add_header(self):
        sh2 = wb[self.sht_name]
        print('请输入表头：')
        date = input()
        price = input()
        sh2.append([date, price])
        
    def data_screen(self):
        sh1 = wb['btc']
        sh2 = wb[self.sht_name]
        print('请输入筛选条件：')
        ent = input()
        if ent == '2013':
            for row in range(2, sh1.max_row+1):
                grade = sh1.cell(row=row, column=1).value
                if grade < '2014':
                    sh2['A' + str(row)] = grade
                    sh2['B' + str(row)] = sh1.cell(row=row, column=2).value
                    print('in sh2', sh2['A'+str(row)].value, sh2['B'+str(row)].value)
        elif ent == '2014':
            for row in range(2, sh1.max_row+1):
                grade = sh1.cell(row=row, column=1).value
                if grade > str(2014) and grade < str(2015):
                    sh2['A' + str(row)] = grade
                    sh2['B' + str(row)] = sh1.cell(row=row, column=2).value
                    print('in sh2', sh2['A'+str(row)].value, sh2['B'+str(row)].value)
        elif ent == '2015':
            for row in range(2, sh1.max_row+1):
                grade = sh1.cell(row=row, column=1).value
                if grade > str(2015) and grade < str(2016):
                    sh2['A' + str(row)] = grade
                    sh2['B' + str(row)] = sh1.cell(row=row, column=2).value
                    print('in sh2', sh2['A'+str(row)].value, sh2['B'+str(row)].value)
        elif ent == '2016':
            for row in range(2, sh1.max_row+1):
                grade = sh1.cell(row=row, column=1).value
                if grade > str(2016) and grade < str(2017):
                    sh2['A' + str(row)] = grade
                    sh2['B' + str(row)] = sh1.cell(row=row, column=2).value
                    print('in sh2', sh2['A'+str(row)].value, sh2['B'+str(row)].value)
        elif ent == '2017':
            for row in range(2, sh1.max_row+1):
                grade = sh1.cell(row=row, column=1).value
                if grade > str(2017) and grade < str(2018):
                    sh2['A' + str(row)] = grade
                    sh2['B' + str(row)] = sh1.cell(row=row, column=2).value
                    print('in sh2', sh2['A'+str(row)].value, sh2['B'+str(row)].value)
        else:
            for row in range(2, sh1.max_row+1):
                grade = sh1.cell(row=row, column=1).value
                if grade > str(2018):
                    sh2['A' + str(row)] = grade
                    sh2['B' + str(row)] = sh1.cell(row=row, column=2).value
                    print('in sh2', sh2['A'+str(row)].value, sh2['B'+str(row)].value)
                    
    def save(self):
        print('文件名')
        name = input()
        wb.save(name)
        
```

```python
fq = Excel_data('From', 'to', 'sht_name')
fq.run_excel()
```
