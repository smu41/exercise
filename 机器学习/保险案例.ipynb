```python
# 建模
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
X = data.iloc[:, 1:]
y = data['resp_flag']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=420)
```

```python
dtc = tree.DecisionTreeClassifier().fit(X_train, y_train)
dtc.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# 尝试使用最小叶节点样本数量和最小分割样本数量进行调参
param_grid = {'min_samples_leaf': list(range(1000, 6000, 100)), 
              'min_samples_split': list(range(4000, 6000, 100))}
gcv = GridSearchCV(tree.DecisionTreeClassifier(), param_grid, cv=5)
gcv.fit(X_train, y_train)
```

```python
dtc = tree.DecisionTreeClassifier(criterion='gini', max_depth=4, 
                                  min_samples_leaf=1500, min_samples_split=4000).fit(X_train, y_train)
dtc.score(X_test, y_test)
```
